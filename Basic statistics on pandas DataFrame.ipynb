Here I am going to show you how to perform some basic statistics on pandas DataFrame. I am using a dataset
that I have downloaded from Kaggle - Climate Change: Earth Surface Temperatures:
https://www.kaggle.com/berkeleyearth/climate-change-earth-surface-temperature-data.

```python
import pandas as pd
```

```python
# I read the CSV file.
temperatures = pd.read_csv("GlobalTemperatures.csv")
temperatures.head()
```

Getting the sum of all the values of a column:
df['column_name'].sum()

```python
temperatures['LandAverageTemperature'].sum()
```

By default, this function will skip the missing values. We can choose to skip or include the missing values by passing the following argument:
skipna=True
skipna=False

```python
# skipna=True should give me the same result not passing this argument at all.
temperatures['LandAverageTemperature'].sum(skipna=True)
```

Finding the artithmetic mean of a particular column in the DataFrame:
df['column_name'].mean()

```python
temperatures['LandAverageTemperature'].mean()
```

Passing the argument of axis=0 returns the mean of every single column in the DataFrame:
df.mean(axis=0)

```python
temperatures.mean(axis=0)
```

Passing the argument of axis=1 will return the mean of every single row in the DataFrame:
df.mean(axis=1)

```python
temperatures.mean(axis=1)
```

Using the describe method to find summary statistics:
df['column_name'].describe()

```python
# Summary statistics of the LandAverageTemperature column in the DataFrame.
temperatures['LandAverageTemperature'].describe()
```

You can also choose specific percentiles to be included in the describe method output by adding the percentiles argument and specifying:
df['column_name'].describe(percentiles=[percentile1, percentile2, percentile3, percentile4]
You can change the number of percentiles you ask for as you please - 4 percentiles are just an example.

```python
# Summary statistics of the LandAverageTemperature column in the DataFrame with different percentiles
temperatures['LandAverageTemperature'].describe(percentiles=[.10, .30, .65, 0.80])
```

If your object is non-numerical, the summary statistics will be sligthly different. They will include the count, frequency, the number of unique values and the top value.
If your object contains both numerical and non-numerical values, the describe method will only include summary statistics of the numerical values.

Counting the number of values in a particular column of the DataFrame:
df['col1'].count()

```python
temperatures['LandAverageTemperature'].count()
```

Finding the maximum value in a particular column of a DataFrame:
df['col1'].max()

```python
temperatures['LandAverageTemperature'].max()
```

Finding the minimum value in a particular column of a DataFrame:
df['column_name'].min()

```python
temperatures['LandAverageTemperature'].min()
```

Finding the median value of a particular column in the DataFrame:
df['column_name'].median()

```python
# Median of the LandAverageTemperature column.
temperatures['LandAverageTemperature'].median()
```

Finding the mode of a particular column in the DataFrame:
df['column_name'].mode()

```python
# Mode of the LandAverageTemperature column.
temperatures['LandAverageTemperature'].mode()
```
