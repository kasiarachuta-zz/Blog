```python
'''In this example, I use the data on calls for service from New Orleans.
This data are available here: https://data.nola.gov/
I use the dataset from 2015. I downloaded the CSV files.
'''
```

```python
# I import pandas and I display the precision of floats to 2 digits.
import pandas as pd
```

```python
# I read the CSV file.
twenty15_df = pd.read_csv("Calls_for_Service_2015.csv")
twenty15_df.head()
```

```python
'''We can see about that some cells have NaN in them. That means that the value is missing.
If you want to check if there is a missing value in the cell, use
df.isnull()
This returns a boolean - True if the value is NaN, False if the value is there.'''
```

```python
twenty15_df.isnull()
```

```python
'''You can just easily get rid of missing values - and drop all the records that have at least one missing value:
df.dropna()'''
```

```python
twenty15_df.dropna()
```

```python
'''You do not have to drop all the records that have at least one missing value. You can choose
to drop the row if there are any missing values or if there are all missing values.
Add how to the argument, for example how to equal to all means that it will drop rows where
all the values are null. How equal to any will drop the row where there is at least one missing value.'''
```

```python
twenty15_df.dropna(how='all')
```

```python
'''You can choose to just drop columns that have missing values by specifying the axis=1:
df.dropna(axis=1)'''
```

```python
twenty15_df.dropna(axis=1)
# Here you can see that TimeDispatch got dropped, whihc is what you would expect - that column has missing values.
```

```python
'''Additionally, I can pick how many data points I want by using
df.dropna(thresh=n)
where n specifies the minimum number of data points we want.'''
```

```python
twenty15_df.dropna(thresh=5)
```

```python
'''We can also fill NaN with some value - for example, we may want to replace it with 0.
df.fillna(value)'''
```

```python
twenty15_df.fillna(0)
```

```python
'''You can modify the existing object in two ways: by saving it again e.g.
twenty15_df = twenty15_df.fillna(0)
or by using inplace and equalling it to True.
twenty15_df.fillna(0, inplace=True)'''
```

```python
twenty15_df.fillna(0, inplace=True)
twenty15_df
```
